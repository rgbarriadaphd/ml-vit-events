```python
import numpy as np
import pandas as pd
from scipy import stats
```

## No balanced classes

```python
df_no_balanced = pd.read_csv('20241007_165447_balanced_classes_False/results_table.csv')
df_no_balanced
```

## Balanced classes

```python
df_balanced = pd.read_csv('20241007_164850_balanced_classes_True/results_table.csv')
df_balanced
```

## ALL setting

```python
setting = 'ALL'
df_balanced_all = df_balanced[df_balanced["setting"] == setting].copy()
df_no_balanced_all = df_no_balanced[df_no_balanced["setting"] == setting].copy()

df_balanced_all['balanced'] = True
df_no_balanced_all['balanced'] = False

merged_df = pd.concat([df_balanced_all, df_no_balanced_all], ignore_index=True)
sorted_df = merged_df.sort_values(by='classifier').drop(['setting'], axis=1)
sorted_df
```

## No RD setting

```python
setting = 'NO RD'
df_balanced_all = df_balanced[df_balanced["setting"] == setting].copy()
df_no_balanced_all = df_no_balanced[df_no_balanced["setting"] == setting].copy()

df_balanced_all['balanced'] = True
df_no_balanced_all['balanced'] = False

merged_df = pd.concat([df_balanced_all, df_no_balanced_all], ignore_index=True)
sorted_df = merged_df.sort_values(by='classifier').drop(['setting'], axis=1)

sorted_df
```

## RD setting

```python
setting = 'RD'
df_balanced_all = df_balanced[df_balanced["setting"] == setting].copy()
df_no_balanced_all = df_no_balanced[df_no_balanced["setting"] == setting].copy()

df_balanced_all['balanced'] = True
df_no_balanced_all['balanced'] = False

merged_df = pd.concat([df_balanced_all, df_no_balanced_all], ignore_index=True)
sorted_df = merged_df.sort_values(by='classifier').drop(['setting'], axis=1)

sorted_df
```

```python
from IPython.display import display, Markdown
latex = sorted_df.drop(['train_control','train_event','test_control','test_event'], axis=1).style.to_latex()
display(Markdown(rf"""{latex}"""))
```
